```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from tabulate import tabulate
```

```python
x1=np.linspace(-10,10,2001)
x2=np.linspace(-15,15,2001)
n=np.array([0.001,0.005,0.01,0.05])
x=np.array([x1,x2]).transpose()
x1,x2=np.meshgrid(x1,x2)
```

```python
def f(x1,x2):
    return 10*x1**2 + 5*x1*x2 + 10*(x2-3)**2
def gradient(x1,x2):
    return np.array([(20*x1 + 5*x2),(5*x1 + 20*(x2-3))])
```

```python
z=f(x1,x2)
```

```python
x1min=np.zeros(4)
x2min=np.zeros(4)
iterations_arr=np.zeros(4)
f_value=np.zeros(4)
```

```python
np.linalg.norm(gradient(x1[0][460],x2[588][0]))
```

```python
for i in np.arange(len(n)):
    iterations=0
    xm=np.array([x1[0][0],x2[0][0]])
    for j in np.arange(len(x1)):
        if(x1[0][j]>=-10 and x1[0][j]<=10):
            x1[0][j]=x1[0][j]+n[i]*x1[0][j]
            for k in np.arange(len(x2)):
                if(x2[k][0]>=-15 and x2[k][0]<=15):
                    x2[k][0]=x2[k][0]+n[i]*x2[k][0]
                    iterations=iterations+1
                    xm=abs(gradient(x1[0][j],x2[k][0]))
                    print(n[i],x1[0][j],x2[k][0],xm,np.linalg.norm(xm))
                    if(np.linalg.norm(xm)<0.001):
                        x1min[i]=x1[0][j]
                        x2min[i]=x2[k][0]
                        iterations_arr[i]=iterations
                        f_value=f(x1[0][j],x2[k][0])
                        break
            if(np.linalg.norm(xm)<0.001):
                break
```

```python
data=list(zip(n,iterations_arr,x1min,x2min,f_value))
header=["Step Size","Iterations","x1","x2","Function Value"]
table=tabulate(data,header)
print(table)
```
