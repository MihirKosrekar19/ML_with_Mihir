```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
def f(x):
    return x*np.exp(-x**2)
```

```python
def df(x):
    return np.exp(-x**2)-2*x**2*np.exp(-x**2)
```

```python
def gradient_descent(x_initial, step_size, thresold):
    x=x_initial
    iterations=0
    while True:
        gradient=df(x)
        if np.linalg.norm(gradient)<thresold:
            break
        x=x-step_size*gradient
        iterations+=1
    return x, f(x), iterations
```

```python
x=np.linspace(-np.sqrt(1.5),0,100)
x_initial=x[0]
thresold=0.001
step_sizes=[0.001,0.005,0.01,0.05]
minima_list=[]
minima_value_list=[]
iterations_list=[]
```

```python
for step_size in step_sizes:
    minima, minima_value, iterations=gradient_descent(x_initial, step_size, thresold)
    minima_list.append(minima)
    minima_value_list.append(minima_value)
    iterations_list.append(iterations)
```

```python
plt.plot(x,f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('plot of function f(x)=x*exp(-x**2)')
plt.grid()
```

```python
data=list(zip(step_sizes,iterations_list,minima_list,minima_value_list))
head=['Step Size','Iteration','Minima at','Minima Value']
table=tabulate(data,head)
print(table)
```
